```python
size_data = pd.read_csv('https://raw.githubusercontent.com/eccho03/datamining_project_clothes/refs/heads/main/apriori_result.csv')
```

```python
size_data
```

```python
allowed_sizes = ['XL이상', 'L', 'M', 'S', 'XS이하']
size_data = size_data[size_data['size'].isin(allowed_sizes)]
```

```python
sampled_df = personal_data.sample(n=200, random_state=64)
sampled_df['predict_size'] = ''
```

```python
for index, row in sampled_df.iterrows():
    real_size = row['size']
    if 'XL' in real_size:
        sampled_df.loc[index, 'size'] = 'XL이상'
    elif 'XS' in real_size:
        sampled_df.loc[index, 'size'] = 'XS이하'
```

```python
sampled_df.head(20)
```

```python
def get_size_from_height_weight(height, weight):
    size_rows = size_data[
        ((size_data['min_height'].isna()) | (size_data['min_height'] <= height)) &
        ((size_data['max_height'].isna()) | (size_data['max_height'] >= height)) &
        ((size_data['min_weight'].isna()) | (size_data['min_weight'] <= weight)) &
        ((size_data['max_weight'].isna()) | (size_data['max_weight'] >= weight))
    ]
    if not size_rows.empty:
        # Confidence가 가장 높은 사이즈 선택
        best_size_row = size_rows.loc[size_rows['confidence'].idxmax()]
        return best_size_row['size']
    else:
        return None
```

```python
import re
```

```python
def remove_korean(text):
    # 한글 제거: 한글 범위 [가-힣]을 공백으로 대체
    return re.sub(r'[가-힣]', '', text)
```

```python
for index, row in sampled_df.iterrows():
    user_height = row['height']
    user_weight = row['weight']
    user_size = get_size_from_height_weight(user_height, user_weight)
    sampled_df.loc[index, 'predict_size'] = user_size
```

```python
size_list=['XS', 'S', 'M', 'L', 'XL']
```

```python
TP={'XS' : 0, 'S' : 0, 'M' : 0, "L" : 0, 'XL' : 0}
FP={'XS' : 0, 'S' : 0, 'M' : 0, "L" : 0, 'XL' : 0}
TN={'XS' : 0, 'S' : 0, 'M' : 0, "L" : 0, 'XL' : 0}
FN={'XS' : 0, 'S' : 0, 'M' : 0, "L" : 0, 'XL' : 0}
#[XS, S, M, L, XL]
for index, row in sampled_df.iterrows():
    real_size = row['size']
    predict_size = row['predict_size']
    real_size=remove_korean(real_size)
    predict_size=remove_korean(predict_size)
    #True
    if real_size==predict_size:
        TP[predict_size]+=1
        for size in size_list:
            if size==predict_size:
                continue
            TN[size]+=1

    #False
    else:
        FP[predict_size]+=1
        FN[real_size]+=1
        for size in size_list:
            if size==predict_size or size==real_size:
                continue
            TN[size]+=1
```

```python
precision = {'XS' : 0, 'S' : 0, 'M' : 0, "L" : 0, 'XL' : 0}
recall = {'XS' : 0, 'S' : 0, 'M' : 0, "L" : 0, 'XL' : 0}
```

```python
for size in size_list:
    if TP[size]+FP[size]==0:
        precision[size]=0
    else:
        precision[size] = TP[size]/(TP[size]+FP[size])

    if TP[size]+FN[size]==0:
        recall[size]=0
    else:
        recall[size] = TP[size]/(TP[size]+FN[size])
```

```python
f1score = {'XS' : 0, 'S' : 0, 'M' : 0, "L" : 0, 'XL' : 0}
```

```python
for size in size_list:
  if(precision[size]+recall[size]==0): f1score[size]='NA'
  else : f1score[size]=2*(precision[size]*recall[size])/(precision[size]+recall[size])
```
